# Data Cleaning and Preprocessing
This notebook outlines the standard ingestion pipeline to verify data integrity, format time-series fields, and clean outliers.


```python
import os
import pandas as pd
import numpy as np

# Path configuration
DATA_DIR = '../data'
print('Data Directory:', os.path.abspath(DATA_DIR))
```

## Ingesting Datasets & Checking Integrity


```python
weather_df = pd.read_csv(os.path.join(DATA_DIR, 'weather.csv'))
pollution_df = pd.read_csv(os.path.join(DATA_DIR, 'pollution.csv'))
traffic_df = pd.read_csv(os.path.join(DATA_DIR, 'traffic.csv'))

print('Weather shape:', weather_df.shape)
print('Pollution shape:', pollution_df.shape)
print('Traffic shape:', traffic_df.shape)
```

## Checking for Null Values and Data Types


```python
print('Missing Values in Weather:\n', weather_df.isnull().sum())
print('\nMissing Values in Pollution:\n', pollution_df.isnull().sum())
print('\nMissing Values in Traffic:\n', traffic_df.isnull().sum())
```

## Standardizing Date Fields and Column Formats


```python
# Convert Date columns to datetime object
weather_df['Date'] = pd.to_datetime(weather_df['Date'])
pollution_df['Date'] = pd.to_datetime(pollution_df['Date'])
traffic_df['Date'] = pd.to_datetime(traffic_df['Date'])

print('Weather Date type:', weather_df['Date'].dtype)
```

## Final Cleaned Output Validation


```python
print('Weather Summary Stats:\n', weather_df.describe())
print('Ingestion & Cleaning completed successfully!')
```
