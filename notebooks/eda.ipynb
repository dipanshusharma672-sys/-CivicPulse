# Exploratory Data Analysis (EDA)
In this notebook, we explore the correlations between weather conditions, air quality, electricity consumption, and transport ridership.


```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATA_DIR = '../data'
sns.set_theme(style='darkgrid')
```

## Load Data and Merge


```python
weather = pd.read_csv(os.path.join(DATA_DIR, 'weather.csv'))
pollution = pd.read_csv(os.path.join(DATA_DIR, 'pollution.csv'))
energy = pd.read_csv(os.path.join(DATA_DIR, 'energy.csv'))

# Merge datasets on Date and City
merged = pd.merge(weather, pollution, on=['Date', 'City'])
merged = pd.merge(merged, energy, on=['Date', 'City'])
print('Merged Data Columns:', merged.columns)
print('Merged Data Shape:', merged.shape)
```

## Correlation Matrix Analysis


```python
corr_cols = ['Temperature', 'Humidity', 'Rainfall', 'Wind_Speed', 'AQI', 'PM2_5', 'Electricity_Consumption', 'Renewable_Energy_Percent']
corr_matrix = merged[corr_cols].corr()
print('Correlation Matrix:\n', corr_matrix.round(2))
```

## Plotting Correlations


```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Urban indicators Correlation Matrix')
plt.show()
```

## Identifying Outliers / Extremes


```python
extreme_aqi = merged[merged['AQI'] > 150]
print(f'Number of days with AQI > 150 (Unhealthy): {len(extreme_aqi)}')
print(extreme_aqi[['Date', 'City', 'AQI', 'Temperature', 'Wind_Speed']].head())
```
